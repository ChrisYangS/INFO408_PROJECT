```python
from sql_connector import connect_database
from sqlalchemy import text
import pandas as pd
```

### Read Listing.csv file

```python

df_list = pd.read_csv('./data/listings.csv')
```

listing.csv file will be splited into mutiple sql tables to for normalised database principle.

* First, takes subset of listing.csv data which are only for listing information and save to a dataset df_list_header

```python
df_list_header = df_list[['id', 'listing_url', 'scrape_id','name', 'description', 'neighborhood_overview', 'picture_url','host_id','neighbourhood','latitude', 'longitude', 'property_type', 'room_type', 'accommodates','bathrooms', 'bathrooms_text', 'bedrooms', 'beds', 'amenities','instant_bookable','region_id']]
df_list_header.head()
```

* Then, takes subset of listing.csv data which are only for region information and group all columns to have unique conbinations, and save to a dataset df_region

```python
df_region = df_list[['region_id','region_name', 'region_parent_id']].groupby(['region_id','region_name', 'region_parent_id'], as_index = False, dropna=False).count()
df_region.head()
```

* Take subset of listing.csv data which are only for region and group all columns to have unique conbinations, and save to a dataset df_region_parent

```python
df_region_parent = df_list[['region_parent_id', 'region_parent_name']].groupby(['region_parent_id', 'region_parent_name'], as_index = False, dropna=False).count()
df_region_parent.head()
```

* Take subset of listing.csv data which are only for host information and group all columns to have unique combinations, and save to a dataset df_region_parent

```python
df_host_detail = df_list[['host_id', 'host_url', 'host_name', 'host_since', 'host_location', 'host_about','host_thumbnail_url','host_picture_url', 'host_neighbourhood']].groupby(['host_id', 'host_url', 'host_name', 'host_since', 'host_location', 'host_about','host_thumbnail_url','host_picture_url', 'host_neighbourhood'], as_index=False, dropna=False).count()

df_host_detail.head()
```

* Take subset of listing.csv data which are only for host statical information and group all columns to have unique combinations, and save to a dataset df_host_stats

```python
df_host_stats = df_list[['host_id', 'host_response_time', 'host_response_rate','host_acceptance_rate', 'host_is_superhost','host_listings_count','host_total_listings_count', 'host_verifications','host_has_profile_pic', 'host_identity_verified']].groupby(['host_id', 'host_response_time', 'host_response_rate','host_acceptance_rate', 'host_is_superhost','host_listings_count','host_total_listings_count', 'host_verifications','host_has_profile_pic', 'host_identity_verified'], as_index=False, dropna=False).count()
df_host_stats.head()
```

* Take subset of listing.csv data which are only for list statical information regarding pricing and days of stay, and save to a dataset df_listings_stay_stats

```python
df_listings_stay_stats = df_list[['id','last_searched','price','minimum_nights', 'maximum_nights','minimum_minimum_nights','maximum_minimum_nights', 'minimum_maximum_nights','maximum_maximum_nights', 'minimum_nights_avg_ntm','maximum_nights_avg_ntm','has_availability','last_scraped']]
df_listings_stay_stats.head()
```

* Take subset of listing.csv data which are only for list statical information regarding customer reviews, and save to a dataset df_listings_review_stats

```python
df_listings_review_stats  = df_list[['id','number_of_reviews','last_scraped','first_review','last_review', 'review_scores_rating','review_scores_accuracy','review_scores_cleanliness', 'review_scores_checkin','review_scores_communication', 'review_scores_location','review_scores_value','reviews_per_month']]
df_listings_review_stats.head()
```

* Read reviews.csv file and save it to a dataframe called df_reviews. This dataset record a review log from different users on different days for each listings.

```python
df_reviews = pd.read_csv('./data/reviews.csv')
df_reviews.head()
```

* Read calendar.csv file and save it to a dataframe called df_calendar. This dataset record a pricing, availability and number of days of staying per each list.

```python
df_calendar = pd.read_csv('./data/calendar.csv')
df_calendar.head()
```

### Connect to PostgreSQL database and load data.
Once all csv files are read and saved to corresponding dataframe, we start to load them into a postgresql database.

* Firstly, as all sql tables are pre-created in the database, we need to truncate by calling SQL stored procedure called *reset_database()* all table data if there is to avoid SQL constrain setting stops to load data.

```python
conn = connect_database()

## truncate all database tables before loading data
result = conn.execute(text("CALL reset_database()"))

conn.commit()

conn.close()
```

* Then start to load dataframe datasets to their corresponding SQL tables. **Ensure to follow the order to load tables.** 

```python
conn = connect_database()

# Region data
df_region_parent.to_sql('nz_region_parent',index=False,if_exists='append', con=conn)

df_region.to_sql('nz_region',index=False,if_exists='append', con=conn)

#Host data

df_host_stats.to_sql('nz_host_stats',index=False,if_exists='append', con=conn)

df_host_detail.to_sql('nz_host_detail',index=False,if_exists='append', con=conn)

#Listing Stay Stats Data
df_listings_stay_stats.to_sql('nz_listings_stay_stats',index=False,if_exists='append', con=conn)

#Listing review stats data
df_listings_review_stats.to_sql('nz_listings_review_stats',index=False,if_exists='append', con=conn)

#Listing data
df_list_header.to_sql('nz_listings',index=False,if_exists='append', con=conn)

conn.close()
```

Due to calendar data and review data are very huge, to avoid SQL server time-out, we split these two dataset loading in different process.

```python
conn = connect_database()
#Calendar data
df_calendar.to_sql('nz_calendar',index=False,if_exists='append', chunksize=100000, con=conn)
conn.close()
```

```python

conn = connect_database()

#Calendar data
df_reviews.to_sql('nz_reviews',index=False,if_exists='append',chunksize=100000, con=conn)

conn.close()
```
